```python
#NUMPY INDEXING AND SLICING OF MATRICES
```

```python
arr = np.arange(1,31).reshape((6,5))

arr
```

```python
slice = arr[3:,3:]    #FIRST SLICING WILL SLICE THE ROW AND THE SECOND SLICING WILL SLICE THE COLUMN
slice
```

```python
a =  arr[:,2]
a
b = arr[::5,::4]  #finding the corner elements 
b
```

```python
# BOOLEAN INDEXING
```

```python
arr = np.arange(11,21)
arr
```

```python
slice_arr = arr%2 == 0
slice_arr
```

```python
arr = arr[slice_arr]
arr
```

```python
a3 = np.arange(27).reshape(3,3,3)
a3
```

```python
a3[1,1,1]  #FIRST WILL FETCH THE 2D MATRIX AND LAST 2 IS SAME AS LIKE 2D MATRIX INDEXING
```

```python
a3[2,1:,1:]
```

```python
a3[::2,0,::2]   #FETCHING THE CORNER ELEMENTS
```
