## Numpy array vs Python lists

```python
a = [i for i in range(10000000)]
b = [i for i in range(10000000,20000000)]
c = []
import time

start = time.time()
for i in range(len(a)):
    c.append(a[i] + b[i])
print(time.time()-start)
```

```python
# numpy
import numpy as np
a = np.arange(10000000)
b = np.arange(10000000,20000000)
start = time.time()
c = a+b
print(time.time() - start)
```

```python
2.5284461975097656/0.05152726173400879  #NUMPY ARRAY IS ALMOST 50 TIMES FASTER THAN PYTHON LIST
```

## Broadcasting ###how numpy treat different arrays during arithmetic operations

```python
#Broadcasting Rules
'''
1. Make the two arrays have the same number of dimensions.

. If the numbers of dimensions of the two arrays are different, add new dimensions with size 1 to the head of the array with the
smaller dimension.

LK

...

2. Make each dimension of the two arrays the same size.

· If the sizes of each dimension of the two arrays do not match, dimensions with size 1 are stretched to the size of the other
array.
· If there is a dimension whose size is not 1 in either of the two arrays, it cannot be broadcasted, and an error is raised.
'''
```

```python
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)
print(a)
print(b)

print(a+b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(3)
print(a)
print(b)

print(a+b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(16).reshape(4,4)
b = np.arange(4).reshape(2,2)

print(a)
print(b)
print(a+b)
```

## Working with mathematical formulas

```python
a = np.arange(10)
np.sin(a)
```

```python
#Sigmoid
def sigmoid(array):
    return 1/(1+ np.exp(-(array)))
sigmoid(a)
```

```python
#mean squared error
actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)
MSE = np.mean((actual - predicted)**2)
MSE
```

```python
#binary cross entropy
L = -np.mean(actual * np.log(predicted) + (1 - actual) * np.log(1 - predicted))
L
```
